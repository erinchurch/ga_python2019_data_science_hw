Visualization for the summary of FNMA Acqustion Data, by Vintage

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = 'Acq_byVintage.csv'
byVintage = pd.read_csv(file, sep = ',')
byVintage
```

```python
plt.style.available
```

```python
byVintage.set_index('VINTAGE')
```

```python
byVintage.plot(y="Total UPB", x='VINTAGE').set_ylabel('UPB $B')
```

```python
byVintage[["BRW FICO WA","CO BRW FICO WA"]].plot().set_ylabel('Credit Score')
```

```python
plt.style.use("fivethirtyeight") 
byVintage.plot(y="Int Rate WA", x='VINTAGE').set_ylabel('Interest Rate %')
```

```python
plt.style.use('seaborn-dark') 
byVintage.plot(y="Int Rate WA", x='VINTAGE').set_ylabel('Interest Rate %')
```

```python
plt.style.use('seaborn-darkgrid') 
byVintage.plot(y="Int Rate WA", x='VINTAGE').set_ylabel('Interest Rate %')
```

```python
plt.style.use('seaborn-talk') 
byVintage.plot(y="Int Rate WA", x='VINTAGE').set_ylabel('Interest Rate %')
```
